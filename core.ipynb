```python
import pandas as pd
from PIL import Image
import requests
import os
import time, math

img_path="image_db/"

if not os.path.exists(img_path):
    os.makedirs(img_path)
df = pd.read_csv('leopidotera-dk.csv', low_memory=False)
#
# Uncomment below depending how many images you want to download
#
df = df[:100]
```

```python
def imgpath_from_row(row, index):
    extension = row['identifier'].split(".")[-1]
    return f"{img_path}{index}.{extension}"
```

```python
def loading_bar(i, max):
    bl = 50
    chars = math.ceil(i * (bl/max))
    p = math.ceil((i/max)*100)
    print('▮' * chars + '▯' * (bl-chars), str(p) + '%', end='\r')
```

```python
def fetch_images():
    r_count=len(df)
    for index, row in df.iterrows():
        imgpath = imgpath_from_row(row, index)
        if not os.path.exists(imgpath):
            Image.open(requests.get(row['identifier'], stream=True).raw).save(imgpath)
        loading_bar(index, r_count)
```
